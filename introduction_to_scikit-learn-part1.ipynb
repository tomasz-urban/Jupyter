## 0. End-to-end Scikit-Learn workflow

```python
# 1. Get the data ready
import numpy as np
import pandas as pd
heart_disease = pd.read_csv('data/heart-disease.csv')
heart_disease
```

```python
# Create X (features matrix)
X = heart_disease.drop("target", axis=1) 

# Create y (labels)
y= heart_disease["target"]
```

```python
# 2. Choose the right model and hyperparameters
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

# We'll keep the default hyperparameters
clf.get_params()
```

```python
# 3. Fit the model to the training data
from sklearn.model_selection import train_test_split

#Split the data to training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
# Find the patterns in the training data 
clf.fit(X_train, y_train);
```

```python
# Make a prediction
y_preds = clf.predict(X_test)
y_preds
```

```python
y_test
```

```python
# 4. Evaluate the model on the training data and test data
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
# Classifications metrics that compare the test labels to the prediction labels
print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds)
```

```python
# 5. Imporve a model
# Try different amount of n_estimators
np.random.seed(42)
for i in range(10, 100, 10):
    print(f'Trying model with {i} estimators')
    clf = RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
    print(f'Model accuracy on test set: {clf.score(X_test, y_test) * 100:.2f}%') # :.2f - 2 decimal places
    print(" ")
```

```python
# 6. Save a model and load it
import pickle
pickle.dump(clf, open("random_forest_model_1.pkl", "wb")) # wb - write binary
```

```python
# .. loading
loaded_model = pickle.load(open("random_forest_model_1.pkl", "rb")) # rb - read binary
loaded_model.score(X_test, y_test)
```
