```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
raw_data = pd.read_csv('data/dummy_variables.csv')
```

```python
raw_data
```

```python
df = raw_data.copy()
```

```python
df['Attendance'] = df['Attendance'].map({'Yes':1, 'No':0})
```

```python
df
```

```python
df.describe()
```

```python
y = df['GPA']
x1 = df[['SAT', 'Attendance']]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y, x).fit()
results.summary()
```

```python
plt.scatter(df['SAT'], y,c=df['Attendance'], cmap='RdYlGn_r')
yhat_no = 0.6439 + 0.0014*df['SAT']
yhat_yes = 0.6429 + 0.0014*df['SAT'] + 0.2226
fig = plt.plot(df['SAT'], yhat_no, lw=2, c='Green')
fig = plt.plot(df['SAT'], yhat_yes, lw=2, c="Red")
plt.xlabel('SAT', fontsize = 15)
plt.ylabel('GPA', fontsize = 15);
```

```python
plt.scatter(df['SAT'], y,c=df['Attendance'], cmap='RdYlGn_r')
yhat_no = 0.6439 + 0.0014*df['SAT']
yhat_yes = 0.6429 + 0.0014*df['SAT'] + 0.2226
yhat = 0.0017*df['SAT'] + 0.275 
fig = plt.plot(df['SAT'], yhat_no, lw=2, c='Green')
fig = plt.plot(df['SAT'], yhat_yes, lw=2, c="Red")
fig = plt.plot(df['SAT'], yhat, lw=2, c="Blue")
plt.xlabel('SAT', fontsize = 15)
plt.ylabel('GPA', fontsize = 15);
```
