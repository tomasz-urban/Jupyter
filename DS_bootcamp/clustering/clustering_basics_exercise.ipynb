## Import the relevant libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
sns.set()
```

## Load the data

Load data from the csv file: <i> 'clustering_basics_exercise.csv'</i>.


```python
data = pd.read_csv('data/clustering_basics_exercise.csv')
data
```

## Plot the data

Plot the <i>'Longtitude'</i> and <i>'Latitude'</i> columns. 

```python
plt.scatter(data['Longitude'], data['Latitude'])
plt.show()
```

## Select the features

Create a copy of that data and remove all parameters apart from <i>Longitude</i> and <i>Latitude</i>.

```python
df_clusters = data.copy()
```

```python
x = df_clusters.iloc[:,1:3]
x
```

## Clustering

Assume there are only two clusters. 

```python
kmeans = KMeans(2)
```

```python
kmeans.fit(x)
```

### Clustering Resutls

```python
clusters = kmeans.fit_predict(x)
clusters
```

```python
df_clusters['Clusters'] = clusters
df_clusters
```

Did you remember to use the <i> c </i> parameter to separate the data by the clusters we defined?

```python
plt.scatter(df_clusters['Longitude'], df_clusters['Latitude'],c=df_clusters['Clusters'], cmap = 'rainbow')
plt.show()
```
