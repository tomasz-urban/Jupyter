## Matplotlib example workflow

```python
# 0. Import matplotlib and get it ready for plotting in Jupyter
# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) # figsize - dimension af a plot (width, height)

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
       xlabel = "x-axis",
       ylabel = "y-axis")

# 5. Save and show (save the whole figure)
fig.savefig("sample-plot.png")
```

```python
# Create a bar plot from dictionary

nut_butter_prices = {"Almond butter": 10,
                    "Peanut butter": 8,
                    "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Butter store",
      ylabel = "Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Create a histogram

x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# Make subplots - two options
```

```python
# 1st option
fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows = 2, ncols=2, figsize=(10, 5))

# Plot to each different axis
ax0.plot(x, x/2);
ax1.scatter(np.random.random(10), np.random.random(10));
ax2.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax3.hist(np.random.randn(1000));
```

```python
# 2nd option
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000));
```

### Which should be used? (pyplot vs matplotlib OO method)
* When plotting something quickly, okay to use the pyplot method
* When plotting something more advanced, use the OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target");
```

```python
# OO method mixed with pyplot method
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
# ax.set_xlim([45, 100]) # limitations of ax's
```

```python
## OO method from scratch
fix, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x= over_50["age"],
                     y = over_50["chol"],
                     c= over_50["target"]);
# Customize the plot
ax.set(title="Heart disease and cholesterol levels",
      xlabel = "Age",
      ylabel = "Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle='--');
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10))

# Add data to ax0
scatter = ax0.scatter(x = over_50["age"],
                      y = over_50["chol"],
                      c = over_50["target"])

# Customize ax0
ax0.set(title="Heart disease and cholesterol levels",
        xlabel = "Age",
        ylabel = "Cholesterol");

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");

# Add data to ax1
scatter = ax1.scatter(x = over_50["age"],
                      y = over_50["thalach"],
                      c = over_50["target"])

# Customize ax1
ax1.set(title = "Heart disease and max heart rate",
       xlabel = "Age",
       ylabel = "Max heart rate");

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")
# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle = '--');
```

```python
# The same code as above but with shared x ax

# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex = True)  # Here added sharex

# Add data to ax0
scatter = ax0.scatter(x = over_50["age"],
                      y = over_50["chol"],
                      c = over_50["target"])

# Customize ax0
ax0.set(title="Heart disease and cholesterol levels",
#       xlabel = "Age",                          # And we remove here (and we get shared "x" ax)
        ylabel = "Cholesterol");

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");

# Add data to ax1
scatter = ax1.scatter(x = over_50["age"],
                      y = over_50["thalach"],
                      c = over_50["target"])

# Customize ax1
ax1.set(title = "Heart disease and max heart rate",
       xlabel = "Age",
       ylabel = "Max heart rate");

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")
# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle = '--');

# Add a title to the figure
fig.suptitle("Heart disease analysis", fontsize=16, fontweight="bold");
```

```python
# Customizing colors (plot used before)

plt.style.use('seaborn-whitegrid')

## OO method from scratch
fix, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x= over_50["age"],
                     y = over_50["chol"],
                     c= over_50["target"],
                     cmap="winter"); # this changes the colour scheme (check matplotlib colours tutorial)
# Customize the plot
ax.set(title="Heart disease and cholesterol levels",
      xlabel = "Age",
      ylabel = "Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle='--');
```

```python
# Customizing y and x axis limitations (on the plot used above)

# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex = True)  # Here added sharex

# Add data to ax0
scatter = ax0.scatter(x = over_50["age"],
                      y = over_50["chol"],
                      c = over_50["target"],
                      cmap = "winter")

# Customize ax0
ax0.set(title="Heart disease and cholesterol levels",
#       xlabel = "Age",                          # And we remove here (and we get shared "x" ax)
        ylabel = "Cholesterol");
# Change the x axis limits
ax0.set_xlim([50, 80])

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");

# Add data to ax1
scatter = ax1.scatter(x = over_50["age"],
                      y = over_50["thalach"],
                      c = over_50["target"],
                      cmap = "winter")

# Customize ax1
ax1.set(title = "Heart disease and max heart rate",
       xlabel = "Age",
       ylabel = "Max heart rate");
# Change ax1 x axis limits
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")
# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle = '--');

# Add a title to the figure
fig.suptitle("Heart disease analysis", fontsize=16, fontweight="bold");
```
